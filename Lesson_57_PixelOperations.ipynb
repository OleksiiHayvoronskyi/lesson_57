```python
# Required libraries.
from PIL import Image
```

```python
# Open an image from a disk.
my_img = Image.open("cat.jpg")
print('Original image')
my_img.show()
```

```python
print('Characteristics of the image:')
print(my_img.format, my_img.size, my_img.mode)
```

### Change rotate the image

```python
# Set the resize method to resize the image.
# In this case I don't want to change a size of the image.
img_rotate = my_img.resize((500, 375))

# Set the rotate method  to transform image to 180 degrees.
img_rotate = my_img.rotate(180)

print('The image has been rotated to 180 degrees')
img_rotate
```
