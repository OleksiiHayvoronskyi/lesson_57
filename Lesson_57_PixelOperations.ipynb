```python
# Required libraries.
from PIL import Image
from PIL import ImageEnhance

from colorama import Fore
```

## Image Open

```python
# Path to an image for further analysis.
path = (r'E:\Python DEVELOPER\StartUpAcademy\lectures\Лекція 57. Піксельні операції\image\cat.jpg')

# Open an image from a disk.
img_origin = Image.open(path)
print(Fore.RED + '\033[1m' + 'Original image')
img_origin
```

```python
# Saving the original image into a file 'img_origin.png'.
img_origin.save('img_origin.png')
print('The image has been successfully saved as a file')
```

```python
print('Characteristics of the original image:')
print(img_origin.format, img_origin.size, img_origin.mode)
```

## Image Contrast

```python
# Creating an object of the Contrast method.
enhancer = ImageEnhance.Contrast(img_origin)

'''Parameters: factor – a floating point value controlling the enhancement.
A factor of 0.0 gives a solid grey image. A factor of 1.0 gives the original image.'''

# Showing a resultant image.
img_contr = enhancer.enhance(factor=3.0)
print(Fore.GREEN + '\033[1m' + 'High Contrast: factor is 3.0')
img_contr
```

```python
# Saving the image into a file 'img_contr.png'.
img_contr.save('img_contr.png')
print('The image has been successfully saved as a file')
```

## Image Brightness

```python
# Creating an object of the Brightness method.
enhancer = ImageEnhance.Brightness(img_origin)

# Showing a resultant image.
img_bright = enhancer.enhance(factor=2.0)
print(Fore.CYAN + '\033[1m' + 'High brightness: factor is 2.0')
img_bright
```

```python
# Saving the image into a file 'img_bright.png'.
img_bright.save('img_bright.png')
print('The image has been successfully saved as a file')
```

## Color Image (saturation)

```python
# Creating na object of the Color method.
enhancer = ImageEnhance.Color(img_origin)

# Showing a resultant image.
img_color = enhancer.enhance(factor=5)
print(Fore.YELLOW + '\033[1m' + 'High brightness: factor is 5.0')
img_color
```

```python
# Saving the image into a file 'img_color.png'.
img_color.save('img_color.png')
print('The image has been successfully saved as a file')
```

## Image resize

```python
# Creating na object of the Resize method.
img_resize = img_origin.resize((600, 500))
print(Fore.BLUE + '\033[1m' + 'The image has been resized from 500 x 375 to 600 x 500')
img_resize
```

```python
# Saving the image into a file 'img_resize.png'.
img_resize.save('img_resize.png')
print('The image has been successfully saved as a file')
```

## Image Cropping

```python
# Setting parameters for cropped image:
# left, upper, right, lower
area = (30, 30, 435, 340)

# Creating na object of the Crop method, and put the parameters.
img_crop = img_origin.crop(area)
print(Fore.RED + '\033[1m' + 'The image has been cropped.')
print(Fore.YELLOW + 'Cropped from each side: Left: 30, Upper: 30, Right: 435, Lower: 340')
img_crop
```

```python
print('Image size before cropped:', img_origin.size)
print('Image size after cropped:', img_crop.size)
```

```python
# Saving the image into a file 'img_crop.png'.
img_crop.save('img_crop.png')
print('The image has been successfully saved as a file')
```

## Image rotate

```python
# Creating na object of the Rotate method.

# Transforming the image to 180 degrees.
img_rotate = img_origin.rotate(180)
print(Fore.MAGENTA + '\033[1m' + 'The image has been rotated to 180 degrees')
img_rotate
```

```python
# Saving the image into a file 'img_rotate.png'.
img_rotate.save('img_rotate.png')
print('The image has been successfully saved as a file')
```
